# TASK:  TO PREDICT THE PERCENTAGE OF A STUDENT BASED ON THE NUMBER OF STUDY HOURS 
   ## (Prediction using supervised ML)

## 1.Importing the Required Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## 2.Reading the Data

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data.head(10)
```

## 3.Visualization and Analysis of Data

```python
data.info()
```

```python
data.describe()
```

```python
sns.heatmap(data.corr(), annot = True, cmap = 'coolwarm', linewidth = 1)
```

```python
#Plotting the distribution of scores
data.plot(x = 'Hours',y = 'Scores', style = 'o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours')
plt.ylabel('Percentage')
plt.show()
```

## 4.Preparing the Data

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## 5.Training the Algorithm

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## 6.Plotting the Line of Regression

```python
#Plotting the Regresion line
line = regressor.coef_*x+regressor.intercept_

#Plotting for the test data
plt.scatter(x,y)
plt.plot(x, line, color = 'Red')
plt.show()
```

## 7.Making Predictions

```python
#Testing Data- in Hours
print(x_test)
#Predicting the Scores
y_pred = regressor.predict(x_test)
```

## 8.Comparing Actual vs Predicted

```python
#Comparing Actual vs Predicted
data1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
data1
```

## 9.Testing with own Data

```python
hours = 9.25
test = np.array([hours])
test = test.reshape(-1,1)
ownpred = regressor.predict(test)
print("No. of hours = {}".format(hours))
print("Predicted score = {}".format(ownpred[0]))
```

## 10.Evaluating the Model

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
